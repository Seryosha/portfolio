### Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2. Напишите код, который проверяет какая из этих строк длиннее.

#### _Примеры работы программы:_
    
    phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
    phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
    Результат:
    Фраза 1 длиннее фразы 2

    phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
    phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
    Результат:
    Фраза 2 длиннее фразы 1

    phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
    phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
    Результат:
    Фразы равной длины

<div class="alert alert-block alert-warning">
Так "фраза" или "строка"?

```python
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
```

```python
def length_comparison_symbols (phrase_1, phrase_2):
# функция принимает в качестве аргументов две строки и сравнивает количество символов в этих строках
    indicator = len(phrase_1) - len(phrase_2)
# так мы вычислим длину каждой фразы только один раз
    if indicator == 0:
        answer = 'Результат:\nФразы равной длины'
    else:
        if indicator > 0:
            short_string = 1; long_string = 2
        else:
            short_string = 2; long_string = 1
# длинновато, в других заданиях так больше делать не станем
        answer = 'Результат:\nФраза ' + str(long_string) + ' длинее фразы ' + str(short_string)
    return print(answer)
```

```python
length_comparison_symbols(phrase_2, phrase_1)
```

```python
def length_comparison_word (phrase_1, phrase_2):
# функция принимает в качестве аргументов две строки и сравнивает в этих строках число слов, разделённых пробелами, в остальном всё то же самое
    indicator = len(phrase_1.split()) - len(phrase_2.split())
    if indicator == 0:
        answer = 'Результат:\nФразы равной длины'
    else:
        if indicator > 0:
            short_string = 1; long_string = 2
        else:
            short_string = 2; long_string = 1
        answer = 'Результат:\nФраза ' + str(long_string) + ' длинее фразы ' + str(short_string)
    return print(answer)
```

#### _Примеры работы программы:_

    year = 2020
    Результат:
    Високосный год

    year = 2019
    Результат:
    Обычный год

> Високосным годом является тот год, который нацело делится на 4, кроме столетий (исключением являются столетия делящиеся на 400).

```python
# перехватим попытку юзера ввести вместо цифр что-то непотребное
try: 
    year = int(input('Введите год в формате четырёхзначного числа:'))
# не надо нам нечетырёхзначного года!
    if (len(str(year)) != 4):
        print('"Год в формате четырёхзначного числа" это когда четыре цифры, а не вот это вот всё.')
    else:
# собственно проверка на високосность
        if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0: 
            print (f'year = {year}\nРезультат:\nВисокосный год')
        else:
            print (f'year = {year}\nРезультат:\nОбычный год')
except Exception:
    print('Число должно состоять из только лишь цифр. Цифры это примерно такое: 5, 2, 4... ну, максимум 9.')
```

```python
# необходимые функции

def month_number (month):
    """
    Функция возвращает номер месяца в соответствии с его названием. Ограничение на использование именительного падежа обусловлено тем, что
    иначе код неоправданно усложняется. Особое огорчение вызывает попытка юзера написать что-то вроде "мартабря", то есть когда первые 
    буквы в названии месяца верные, а дальше абракадабра.
    "Зачем делать сложным то, что проще простого?"
    """

    month_disc = {
                'январь': 1,
                'февраль': 2,
                'март': 3,
                'апрель': 4,
                'май': 5,
                'июнь': 6,
                'июль': 7,
                'август': 8,
                'сентябрь': 9,
                'октябрь': 10,
                'ноябрь': 11,
                'декабрь': 12
                }
    try:
        month_numder = month_disc[month.lower()] # март это март и Март это март
        return month_numder
    except Exception:
        print(' Это не название месяца или название не в именительном падеже!')

def zodiac (number_day):
    """
    Функция возвращает знак зодиака, соответствующий порядковому номеру дня года. При этом исходим из предположения, что все люди родились в одном,
    допустим, 1900 году, и на результат это не повлияет, я гарантирую.
    """
    zodiac_list = [[1, 20, 'Козерог'],
                   [21, 49, 'Водолей'],
                   [50, 79, 'Рыбы'],
                   [80, 110, 'Овен'],
                   [111, 140, 'Телец'],
                   [141, 172, 'Близнецы'],
                   [173, 203, 'Рак'],
                   [204, 234, 'Лев'],
                   [235, 266, 'Дева'],
                   [267, 296, 'Весы'],
                   [297, 326, 'Скорпион'],
                   [327, 355, 'Стрелец'],
                   [356, 365, 'Козерог']]
    for i in zodiac_list:
        if number_day >= i[0] and day <= i[1]:
            zodiac_sign = i[2]
    return zodiac_sign
```

```python
# не обошлось таки без импорта библиотеки
import datetime 
# исключим всякие 32 мая
try: 
    day = int(input(' Введите число месяца:\n'))
# про необходимость именительного падежа объяснено в комментариях к функции month_number
    month = input(' Введите месяц в именительном падеже:\n')
# собственно, год может быть любой. Нужен только лишь для корректной работы datetime.date()
    year = 1900
# порядковый номер дня рождения с начала 1900 года
    number_day = int(datetime.date(year, month_number (month), day).strftime("%j"))
# определяем знак Зодиака с использованием ранее описанной функции
    zodiac_sign = zodiac (number_day)
    print (f' Результат:\n Ваш знак зодиака: {zodiac_sign}')
except Exception:
    print(' Не бывает такой даты!')
```

### Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах).

#### Используйте следующие правила:

* если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран “Коробка №1”;
* если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите “Коробка №2”;
* если длина товара больше 2 метров, то выводите “Упаковка для лыж”;
* во всех остальных случаях выводите “Стандартная коробка №3”.

#### _Примеры работы программы:_
    width = 15
    length = 50
    height = 15
    Результат:
    Стандартная коробка №3

    width = 45
    length = 205
    height = 45
    Результат:
    Ищите упаковку для лыж

```python
width = 50
length = 60
height = 15

# длина это не просто то, что юзер записал в length. Длина это наибольшее из трёх измерений
if length >= width and length >= height:
# закинем размеры в список и будем работать с ним
    product_dimensions = [width, length, height]
    if product_dimensions[1] > 200:
        print('Результат:\nУпаковка для лыж')
    elif all([x <= 15 for x in product_dimensions]):
        print('Результат:\nКоробка № 1')
    elif any([x > 15 and x < 50 for x in product_dimensions]):
        print('Результат:\nКоробка № 2')
    else:
        print('Результат:\nСтандартная коробка № 3')
else:
    print('Некорректно указаны габариты товара (длина должна быть наибольшей!)')
```

### Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет “счастливым”.

> Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

#### _Пример работы программы:_
    number = 123456
    Результат:
    Несчастливый билет

    number = 123321
    Результат:
    Счастливый билет

```python
# номер билета может начинаться с 0, потому int сразу не подходит
number = str(input('number = '))
# обработаем исключение, возникающее в случае ввода юзером других символов помимо цифр
try:
# теперь преобразуем строку в список int'ов    
    list_digits = [int(x) for x in str(number)]
# проверим длину номера и не состоит ли он весь из 0    
    if len(list_digits) == 6 and sum(list_digits) != 0:
# проверим билет на счастливость
        answer = 'Счастливый билет' if sum(list_digits[:3]) == sum(list_digits[3:6]) else 'Несчастливый билет'
        print(f'Результат:\n{answer}')
    else:
        print('Билета с таким номером не может быть')
except Exception:
    print('Шутите?')
```

### А давайте здесь уже отклонимся от ТЗ? И сделаем так:

```python
# объясним юзеру, что же от него требуется
print('Введите в любой последовательности одно, два или три размерения фигуры: одно (радиус) если это круг, два (длина и ширина) если прямоугольник (квадрат),\nтри (длины сторон) если треугольник. При отсутствии размерения нажимайте Enter.')
dimension_1 = input('Первое размерение:')
dimension_2 = input('Второе размерение:')
dimension_3 = input('Третье размерение:')
dimensions = [dimension_1, dimension_2, dimension_3]
# доколе можно объяснять, что именно сделано не так? Пусть юзер сам разбирается
try:
# заменим '' на 0    
    for index, item in enumerate(dimensions):
        if item  == '':
            dimensions[index] = 0
# преобразуем тип размерений во float
    for i in range (0, 3): dimensions[i] = float(dimensions[i]) 
# и отсортируем список размерений по возрастанию    
    dimensions.sort()
# проверяем элементы списка, начиная с [0], и делаем соответствубщие выводы о фигуре, рассчитываем площадь
    if dimensions[0] != 0:
        fig = 'треугольника'
# формула Герона же        
        p = sum(dimensions) / 2
        s = round((p * (p - dimensions[0]) * (p - dimensions[1]) * (p - dimensions[2]))**0.5, 2)
    elif dimensions[1] != 0:
        if dimensions[1] == dimensions[2]:
            fig = 'квадрата'
        else:
            fig = 'прямоугольника'
        s = round (dimensions[1] * dimensions[2], 2)
    elif dimensions[2] != 0:
        fig = 'круга'
        s = round (3.14*(dimensions[2]**2), 2)
# если все три размерения 0
    else:
        fig = 'геометрической точки всегда'
        s = 0
    print(f'Площадь {fig} равна {s} кв. у.е.')
except Exception:
    print('Введены некорректные данные')
```
